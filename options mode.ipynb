```
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 1/4 * x**4 + x**2 - 8*x + 12

def df(x):
    return x**3 + 2*x - 8

def d2f(x):
    return 3*x**2 + 2

a = 0
b = 2
epsilon = 0.5

n = int((b - a) / epsilon)
print("n =", n)

x_values = [a + k * (b - a) / n for k in range(n + 1)]
f_values = [f(x) for x in x_values]

print("\nX արժեքներ և f(X)")
for x in x_values:
    print(f"x = {x:.2f}   f(x) = {f(x):.4f}")

min_index = np.argmin(f_values)
x_min = x_values[min_index]
f_min = f_values[min_index]

print("\nԱղյուսակային մինիմում")
print("X_min =", x_min)
print("f_min =", f_min)

x0 = 1.5
for _ in range(20):
    x0 = x0 - df(x0) / d2f(x0)

x_star = x0
f_star = f(x_star)

print("x* ≈", round(x_star, 4))
print("f(x*) ≈", round(f_star, 4))



x_plot = np.linspace(a, b, 400)
y_plot = f(x_plot)

plt.figure()
plt.plot(x_plot, y_plot)
plt.scatter(x_values, f_values)
plt.scatter(x_min, f_min)
plt.scatter(x_star, f_star)

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Graph of f(x)")
plt.show()
```
