```
import numpy as np
import matplotlib.pyplot as plt

# Ֆունկցիան
def f(x):
    return 1/3 * x**4 + 3*x**2 - 12*x + 10

# Առաջին ածանցյալ (f'(x) = 0 գտնելու համար)
def df(x):
    return 4/3*x**3 + 6*x - 12

# Երկրորդ ածանցյալ
def d2f(x):
    return 4*x**2 + 6

```

```
a, b = 0, 3
epsilon = 0.5
n = int((b - a) / epsilon)

# Կետերի ստեղծում
x_values = [a + k * (b - a) / n for k in range(n + 1)]
f_values = [f(x) for x in x_values]

# Արդյունքների տպում աղյուսակի տեսքով
print(f"{'x':>10} | {'f(x)':>10}")
print("-" * 25)
for i in range(len(x_values)):
    print(f"{x_values[i]:10.2f} | {f_values[i]:10.4f}")

# Աղյուսակային մինիմումի որոշում
min_index = np.argmin(f_values)
x_min, f_min = x_values[min_index], f_values[min_index]

print(f"\nԱղյուսակային մինիմում: x = {x_min}")
```

```
x_star = x_min  # Սկսում ենք 1.5-ից
iterations = 10

for i in range(iterations):
    # Նյուտոնի բանաձևը
    x_star = x_star - df(x_star) / d2f(x_star)

print(f"Նյուտոնի մեթոդից հետո x* ≈ {x_star:.4f}")
print(f"Ֆունկցիայի արժեքը այդ կետում f(x*) ≈ {f(x_star):.4f}")

# Սխալանքի հաշվարկ
error = abs(x_star - x_min)
print(f"Սխալանքը |x* - x_min| = {error:.4f}")
```

```
x_plot = np.linspace(a, b, 400)
y_plot = f(x_plot)

plt.figure(figsize=(10, 6))
plt.plot(x_plot, y_plot, label="f(x)", color='black')

# Աղյուսակային կետերը (կապույտ)
plt.scatter(x_values, f_values, color='blue', label="Points")

# Աղյուսակային մինիմումը (կանաչ)
plt.scatter(x_min, f_min, color='green', s=100, label=f"Tabular Min ({x_min})")

# Ճշգրիտ մինիմումը (կարմիր աստղ)
plt.scatter(x_star, f(x_star), color='red', marker='*', s=200, label=f"New Min ({x_star:.3f})")

plt.grid(True, linestyle=':')
plt.legend()
plt.title("Comparison: Tabulation vs New")
plt.show()
```
