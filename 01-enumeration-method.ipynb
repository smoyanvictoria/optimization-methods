```
import numpy as np
import matplotlib.pyplot as plt
```

```
def f(x):
    return 1/3 * x**4 + 3*x**2 - 12*x + 10

def df(x):
    return 4/3*x**3 + 6*x - 12

def d2f(x):
    return 4*x**2 + 6
```

```
a = 0
b = 3
epsilon = 0.5

n = int((b - a) / epsilon)

print("n =", n)
```

```
x_values = [a + k * (b - a) / n for k in range(n + 1)]
f_values = [f(x) for x in x_values]

print("\nX արժեքներ և f(X)")

for x in x_values:
    print(f"x = {x:.2f}   f(x) = {f(x):.4f}")
```

```
min_index = np.argmin(f_values)

x_min = x_values[min_index]
f_min = f_values[min_index]

print("\nԱղյուսակային մինիմում")

print("X_min =", x_min)
print("f_min =", f_min)
```

```
x0 = 1.5

for _ in range(20):
    x0 = x0 - df(x0) / d2f(x0)

x_star = x0
f_star = f(x_star)

print("x* ≈", round(x_star,4))
print("f(x*) ≈", round(f_star,4))
```

```
x_plot = np.linspace(a, b, 400)
y_plot = f(x_plot)

plt.figure()

plt.plot(x_plot, y_plot, label="f(x)")

plt.scatter(x_values, f_values, label="Tabulation points")

plt.scatter(x_min, f_min, color='green', label="Tabular min")

plt.scatter(x_star, f_star, color='red', label="Newton min")

# աղյուսակային մինիմումի գրություն
plt.annotate(f"({x_min:.2f} , {f_min:.2f})",
             (x_min,f_min),
             textcoords="offset points",
             xytext=(5,5))

plt.annotate(f"({x_star:.3f} , {f_star:.3f})",
             (x_star,f_star),
             textcoords="offset points",
             xytext=(5,-15))

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Graph of f(x)")

plt.legend()

plt.show()
```
