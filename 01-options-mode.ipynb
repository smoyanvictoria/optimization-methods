```
import numpy as np
import matplotlib.pyplot as plt

# Ֆունկցիան
def f(x):
    return 1/3 * x**4 + 3*x**2 - 12*x + 10

# Առաջին ածանցյալ (f'(x) = 0 գտնելու համար)
def df(x):
    return 4/3*x**3 + 6*x - 12

# Երկրորդ ածանցյալ
def d2f(x):
    return 4*x**2 + 6

```

```
# Նախնական տվյալներ
a, b = 0, 3
epsilon = 0.5
n = int((b - a) / epsilon)

# 1. Աղյուսակային  մեթոդ
x_values = [a + k * (b - a) / n for k in range(n + 1)]
f_values = [f(x) for x in x_values]

min_index = np.argmin(f_values)
x_min, f_min = x_values[min_index], f_values[min_index]

x_star = 1.5 # Սկզբնական մոտարկում
for _ in range(10):
    x_star = x_star - df(x_star) / d2f(x_star)

f_star = f(x_star)

# Արդյունքների տպում
print(f"n = {n}")
print(f"Աղյուսակային X_min = {x_min}")
print(f"Ճշգրիտ (New) x* ≈ {x_star:.4f}")
print(f"Սխալանք |x* - x_min| = {abs(x_star - x_min):.4f}")
```

```
x_plot = np.linspace(a, b, 400)
y_plot = f(x_plot)

plt.figure(figsize=(10, 6))
plt.plot(x_plot, y_plot, label="f(x)", color='black', linewidth=1.5)

# Աղյուսակային կետերը
plt.scatter(x_values, f_values, color='blue', s=30, label="Tabulation points")

# Աղյուսակային մինիմում (Կանաչ)
plt.scatter(x_min, f_min, color='green', s=100, label=f"Tabular min ({x_min})", zorder=5)

# Նոր մինը (Կարմիր)
plt.scatter(x_star, f_star, color='red', marker='*', s=150, label=f"New min ({x_star:.3f})", zorder=6)

# Գրություններ կետերի վրա
plt.annotate(f"({x_min:.2f}, {f_min:.2f})", (x_min, f_min), textcoords="offset points", xytext=(5, 5), color='green')
plt.annotate(f"({x_star:.3f}, {f_star:.3f})", (x_star, f_star), textcoords="offset points", xytext=(5, -15), color='red')

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Comparison of Tabular and New Minimums")
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
```
