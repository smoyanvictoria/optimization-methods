```
import numpy as np
import plotly.graph_objects as go

# Մեր հիմնական ֆունկցիան
def f(x):
    return (1/3)*x**4 + 3*x**2 - 12*x + 10
```

```
# Տրված տվյալները
a = 0
b = 3
epsilon = 0.5
delta = 0.125

iterations = []
temp_a, temp_b = a, b

# Ալգորիթմի իրականացում
while (temp_b - temp_a) > epsilon:
    mid = (temp_a + temp_b) / 2
    x1 = mid - delta
    x2 = mid + delta

    f1, f2 = f(x1), f(x2)

    # Պահպանում ենք տվյալները գրաֆիկի համար
    iterations.append((temp_a, temp_b, mid, x1, x2, f1, f2))

    if f1 < f2:
        temp_b = x2
    else:
        temp_a = x1

print(f"Հաշվարկն ավարտվեց: Կատարվել է {len(iterations)} իտերացիա:")
print(f"Վերջնական միջակայքը՝ [{temp_a:.4f}, {temp_b:.4f}]")
```

```
# Ֆունկցիայի հարթ կորը գրաֆիկի համար
x_graph = np.linspace(0, 3, 400)
y_graph = f(x_graph)

# Յուրաքանչյուր իտերացիայի համար կառուցում ենք գրաֆիկ
for i, (a_curr, b_curr, mid, x1, x2, f1, f2) in enumerate(iterations):
    fig = go.Figure()

    # Ֆունկցիայի գիծը
    fig.add_trace(go.Scatter(x=x_graph, y=y_graph, mode='lines', name='f(x)', line=dict(color='black')))

    # Կենտրոնական և փորձնական կետերը
    fig.add_trace(go.Scatter(x=[mid], y=[f(mid)], mode='markers', marker=dict(size=10, color='green'), name='Midpoint'))
    fig.add_trace(go.Scatter(x=[x1, x2], y=[f1, f2], mode='markers',
                             marker=dict(size=12, color='red', symbol='x'), name='x1, x2 (δ շեղում)'))

    # Հատվածի սահմանները (a և b)
    fig.add_vline(x=a_curr, line_dash="dash", line_color="blue", annotation_text=f"a={a_curr:.2f}")
    fig.add_vline(x=b_curr, line_dash="dash", line_color="blue", annotation_text=f"b={b_curr:.2f}")

    fig.update_layout(
        title=f"Իտերացիա {i+1} | mid = {mid:.3f} | միջակայք [{a_curr:.3f}, {b_curr:.3f}]",
        xaxis_title="x", yaxis_title="f(x)",
        xaxis=dict(range=[0, 3]), template="plotly_white",
        showlegend=True
    )
    fig.show()
```
