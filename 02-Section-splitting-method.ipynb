```
import numpy as np
import matplotlib.pyplot as plt

# Մեր օպտիմալացման ֆունկցիան
def f(x):
    return (1/3)*x**4 + 3*x**2 - 12*x + 10
```

```
# Նախնական տվյալներ
a_start, b_start = 0, 3
epsilon = 0.5
delta = 0.125

a, b = a_start, b_start
iterations = []

# Ալգորիթմի իրականացում
while (b - a) > epsilon:
    mid = (a + b) / 2
    x1 = mid - delta
    x2 = mid + delta

    f1, f2 = f(x1), f(x2)

    # Պահպանում ենք տվյալները գրաֆիկների համար
    iterations.append({
        'a': a, 'b': b, 'mid': mid,
        'x1': x1, 'x2': x2,
        'f1': f1, 'f2': f2
    })

    # Միջակայքի սեղմում
    if f1 < f2:
        b = x2
    else:
        a = x1

print(f"2. Հաշվարկն ավարտվեց: Կատարվել է {len(iterations)} իտերացիա:")
print(f"Վերջնական մոտավոր մինիմումը x* ≈ {(a + b) / 2:.4f}")
```

```
# Ֆունկցիայի հարթ կորի տվյալները
x_vals = np.linspace(a_start, b_start, 400)
y_vals = f(x_vals)

# Յուրաքանչյուր քայլի համար կառուցում ենք առանձին գրաֆիկ
for i, step in enumerate(iterations):
    plt.figure(figsize=(9, 5))

    # Ֆունկցիայի գիծը
    plt.plot(x_vals, y_vals, color='black', label='f(x)', linewidth=1.5)

    # Midpoint և x1, x2 կետերը
    plt.scatter([step['mid']], [f(step['mid'])], color='green', s=80, label='Midpoint', zorder=5)
    plt.scatter([step['x1'], step['x2']], [step['f1'], step['f2']],
                color='red', marker='x', s=100, label='Trial points (x1, x2)', zorder=5)

    # Միջակայքի սահմանները (a և b)
    plt.axvline(x=step['a'], color='blue', linestyle='--', alpha=0.4, label=f"a = {step['a']:.2f}")
    plt.axvline(x=step['b'], color='blue', linestyle='--', alpha=0.4, label=f"b = {step['b']:.2f}")

    # Ձևավորում
    plt.title(f"Bisection Method - Iteration {i+1}")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.grid(True, linestyle=':', alpha=0.5)
    plt.legend(loc='upper right')

    plt.tight_layout()
    plt.show()
```
